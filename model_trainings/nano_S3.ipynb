```python
from ultralytics import YOLO
from ultralytics import settings
import albumentations as A

settings.update({"wandb": True})

# Define the augmentation pipeline
augmentations = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.2),
    A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.2, rotate_limit=8, border_mode=0, p=0.5),
    A.CLAHE(clip_limit=(1.5, 3), tile_grid_size=(4, 8), p=0.3),   # контраст
    A.RandomGamma(gamma_limit=(80, 120), p=0.3),
    A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.5),
], bbox_params=A.BboxParams(format='yolo')) # Assuming YOLO format for bounding boxes

model = YOLO("yolo12l.pt")
#model = YOLO(r"yolo\yolo12m9\weights\best.pt")
results = model.train(data="yolo_dataset.yaml", epochs=25,
                      project="yolo_nano", name="yolo12n_preprocessed", batch=64, device=0, mosaic=False,
                      cache=True, plots=True, augment=True) #time=3 (limit by training hours), fraction = 0.5 (to train of dataset)
```
