```python
from ultralytics import YOLO
from ultralytics import settings
settings.update({"wandb": True})
model = YOLO("yolo12n.pt")
#model = YOLO(r"yolo\yolo12m9\weights\best.pt")
results = model.train(data="yolo_dataset.yaml", epochs=25,
                      project="yolo_nano", name="yolo12n_preprocessed", batch=8, device=0, mosaic=False,
                      cache=True, plots=True) #time=3 (limit by training hours), fraction = 0.5 (to train of dataset)
```
